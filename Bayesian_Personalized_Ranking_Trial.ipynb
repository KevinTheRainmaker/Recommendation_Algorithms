```python
import os
import pandas as pd
import numpy as np

import scipy.sparse as sps

from sklearn.model_selection import train_test_split
import implicit

path = '/content/drive/MyDrive/data/movielens'
ratings_df = pd.read_csv(os.path.join(path, 'ratings.csv'), encoding='utf-8')

train_df, test_df = train_test_split(ratings_df, test_size=0.2, random_state=1234)

sparse_matrix = train_df.groupby('movieId').apply(lambda x : pd.Series(x['rating'].values, index=x['userId'])).unstack()
sparse_matrix.index.name = 'movieId'

sparse_matrix_processed = sps.csr_matrix(sparse_matrix)

model = implicit.cpu.bpr.BayesianPersonalizedRanking(factors=100, learning_rate=0.01, random_state=1107)

model.fit(item_users=sparse_matrix_processed, show_progress=True)

# recommend items for a user
userid = 10
user_items = sparse_matrix_processed.tocsr()
results = model.recommend(userid, user_items)
results = [list(results[x]) for x in range(len(results))]
for result in results:
  print(result[0])
```
