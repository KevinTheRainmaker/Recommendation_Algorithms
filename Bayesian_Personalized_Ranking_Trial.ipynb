```python
import os
import pandas as pd
import numpy as np

import scipy.sparse as sps

import implicit

path = '/content/drive/MyDrive/data'
sparse_matrix = pd.read_csv(os.path.join(path, 'implicit_trial.csv'), encoding='utf-8')
sparse_matrix.fillna(0, inplace=True)
sparse_matrix_processed = sps.csr_matrix(sparse_matrix)

model = implicit.cpu.bpr.BayesianPersonalizedRanking(factors=100, learning_rate=0.01, random_state=1107)

model.fit(item_users=sparse_matrix_processed, show_progress=True)

# recommend items for a user
userids = [x for x in range(1,2)]
for userid in userids:
  user_items = sparse_matrix_processed.tocsr()
  results = model.recommend(userid, user_items, N=5)
  results = [list(results[x]) for x in range(len(results))]
  print(f'user: No.{userid}')
  for result in results:
    print(result[0])
```
